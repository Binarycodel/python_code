# Simple Neural Network and LSTM Model.....

```python
# word lenth for each coloum
import numpy as np
w_count = []
def get_word_lenght(text):
    w_count.append(len(text))
    return len(text)
    
dataset['word_count'] = dataset['clean_text'].apply(lambda x:get_word_lenght(x))
max_val = dataset['word_count'].max()
print('MEAN_VAL ', max_val)
w = np.array(w_count).mean()
w
```

```python
dataset.head()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from sklearn import model_selection
```

```python
label_encode = LabelEncoder()
sentence = dataset['clean_text']
label = label_encode.fit_transform(dataset['category'])
label
```

```python
xtrain, xtest, ytrain, ytest = model_selection.train_test_split(sentence, label, test_size=.20)
```

```python
print('X train' , len(xtrain))
print('Y train' , len(ytrain))
print('X test' , len(xtest))
print('Y test' , len(ytest))
```

```python
# tokenizaing text ...
tokenizer = Tokenizer(num_words=25000, oov_token="<OOV>")
tokenizer.fit_on_texts(xtrain)
w_index = tokenizer.word_index
print('vocab_size =================>>>>>>>>>>>>  ' , len(w_index))
print(w_index)
```

```python
xtrain_seq = tokenizer.texts_to_sequences(xtrain)
xtest_seq = tokenizer.texts_to_sequences(xtest)
# print(xtest_seq[:3])
```

```python
max_val = 2000
xtrain_pad = pad_sequences(xtrain_seq, maxlen=max_val)
xtest_pad = pad_sequences(xtest_seq, maxlen=max_val)
print('x train pad shape', xtrain_pad.shape)
print('x test pad shape', xtest_pad.shape)
```

```python
vocab_size = 25000
embedding_dim = 100
pad_length =2000

model = tf.keras.Sequential([
    keras.layers.Embedding(vocab_size, embedding_dim, input_length=pad_length),
    keras.layers.GlobalAveragePooling1D(),
    keras.layers.Dense(500, activation='relu'),
    keras.layers.Dense(5, activation='sigmoid')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'] )
model.summary()
```

```python
model.fit(xtrain_pad, ytrain , epochs=10, validation_data=(xtest_pad, ytest), verbose=1)
```

```python
model.evaluate(xtest_pad, ytest)
```

# WORD EMBEDDING...... TRAINING BBC NEW WORD EMBEDDING.......

```python
import gensim
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('bbc_text.csv')
```

```python
data.head()
```

```python
# word_tokenize(data['text'][0])
import string
string.punctuation
```

```python
pt = PorterStemmer()

def simple_tokenize(doc):
    token = word_tokenize(doc) 
    stemed = [d for d in token if not d in string.punctuation]
    return stemed

data['cl_text']  = data['text'].apply(lambda x: simple_tokenize(x))
```

```python
train_doc = data['cl_text']
label_encode = LabelEncoder()
y_data = label_encode.fit_transform(data['category'])
y_data
```

```python
# TRIANING THE ACTUAL WORD TO VEC....
```

```python
vec_model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)

vec_model
```

```python
# building the vocabulary..... 
vec_model.build_vocab(train_doc, progress_per=100)
```

```python
# vec_model.wv.key_to_index /// for printing worda and it uniques index. 
# vec_model.wv.index_to_key /// for printing unique words in the vocab
```

```python
vec_model.train(train_doc, total_examples=vec_model.corpus_count, epochs=vec_model.epochs)
```

```python
vec_model.wv['computer']
```

```python
# SAVING THE MODEL....
vec_model.save('bbc_new_word2vec.model')
print('model saved......')
```

```python
vec_model.wv.most_similar('computer')
```

```python
import gensim
vec_model = gensim.models.Word2Vec.load('bbc_new_word2vec.model')
vec_model
```

```python
from sklearn import model_selection

X_train, X_test, y_train, y_test  = model_selection.train_test_split(train_doc, y_data,  test_size=.20)
```

```python
print('X train' , len(X_train))
print('Y train' , len(y_train))
print('X test' , len(X_test))
print('Y test' , len(y_test))
```

```python
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
num_word = 35000
# token_word = Tokenizer(num_words=num_word, oov_token="<OOV>")
token_word = Tokenizer(num_words=num_word)

token_word.fit_on_texts(train_doc)
vocab_size = len(token_word.word_index)
# token_word.word_index
vocab_size
```

```python
# convert to train and test data sequence 

seq_x_train = token_word.texts_to_sequences(X_train)
seq_x_test = token_word.texts_to_sequences(X_test)

# pad the sequence 
pad_size = len(train_doc.max())
padded_train = pad_sequences(seq_x_train, maxlen=pad_size, padding='post')
padded_test = pad_sequences(seq_x_test, maxlen=pad_size, padding='post')

padded_test.shape
```

```python
# padded_train[:2]
```

```python
# CREATED A WEIGHTED MATRIX USING A PRETRAINED WORD2VEC MODEL
```

```python
# creating an empty weighted matric with zeros...
import numpy as np

vocab_size = len(token_word.word_index)
dimension = 300
weight_matrix = np.zeros((vocab_size, dimension))
weight_matrix.shape
```

```python
# vocab = model.wv.key_to_index
vocab = token_word.word_index
total_vocab = 35000
for word, index in token_word.word_index.items():
#     print(word, (index-1))
    if index < vocab_size: 
        if word in vec_model.wv.index_to_key:
            weight_matrix[index] = vec_model.wv[word]
#             print(len(weight_matrix[index-1]))
        else: 
            weight_matrix[index] = np.zeros(100)

print(weight_matrix.shape)
print('weighted matrix complete....')
```

```python
for w, v in vec_model.wv.key_to_index.items():
    if v == 10:
        break
#     print("{}:{}".format(w, v))
# vec_model.wv.index_to_key
# token_word.word_index
```

```python
weight_matrix[1] == vec_model.wv['the']
```

```python
# lstm model using unsupervise word embedding 
import tensorflow as tf
import tensorflow.keras

vocab_size = len(token_word.word_index)
dimension = 100
pad_length = len(train_doc.max())
classes = 5

lstm_model = keras.Sequential([
    keras.layers.Embedding(vocab_size, dimension, input_length=pad_length, weights=[weight_matrix], trainable=False),
    keras.layers.LSTM(100),
    
#     keras.layers.LSTM(200),
#   dense layer 
#     keras.layers.Dense(200, activation='relu'),
    keras.layers.Dense(classes, activation="sigmoid")
])


lstm_model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

lstm_model.summary()
```

```python
epoch = 10
lstm_model.fit(padded_train, y_train, epochs=3, validation_data=(padded_test, y_test), verbose=1)
```
