```python
new_data = data[['Product Group', 'Product Name', 'Presentation', 'Active Ingredent',
       'ApplicantName', 'Country', 'Manufacturer']]
new_data
```

```python
# filtering nigeria made drug 
```

```python
niger_drug = new_data[(new_data['Product Group']=='DRUG') & (new_data["Country"]=="NIGERIA" ) ]
```

```python
niger_drug[niger_drug.columns[:3]].values
```

```python
niger_drug.columns
```

```python
tem = niger_drug

tem[['drug_name', 'medicine_type', 'ingredient', 'address', 'manufacturer' ]] = niger_drug[['Product Name', 'Presentation', 'Active Ingredent', 'ApplicantName', 'Manufacturer']]
```

```python
tem.drop(columns=['Product Name', 'Presentation', 'Active Ingredent', 'ApplicantName', 'Manufacturer'], inplace=True)
niger_drug = tem
```

```python
niger_drug.drop(columns=['Product Group', 'Country'], inplace=True)
```
