```python
import pandas as pd
import numpy as np
dataset = pd.read_csv('hepatitis_datapoint.data', header=None)
dataset.head()
```

```python
dataset.replace("?", np.NaN, inplace=True )
dataset.head()
```

```python
row , columns = dataset.shape
print(f'contains {row} datapoint and {columns} predictive features')
```

```python
column = ['label' , 'age' , 'sex', 'steroid', 'antivirals', 'fatigue' , 'malaise', 'anorexia', 'liver big', 'liver firm', 'spleen palpable', 'spider', 'ascites' , 'varices', 'bilirubin', 'alk phosphate' , 'sgot', 'albumin', 'protime', 'histology' ]
len(column)
dataset.columns = column
dataset.head()
```

```python
ml_dataset = dataset
text_dataset = dataset
```

```python
text_dataset.dropna()
text_dataset.shape
```

```python
mini_dataset = text_dataset[['age','sex',  'antivirals', 'liver big','bilirubin', 'albumin']]
mini_dataset.head()
```

```python
md = mini_dataset
md.fillna(method='ffill', inplace=True)
md['albumin'].replace('nan','0', inplace=True)


md['albumin'] = md['albumin'].astype(str).astype(float)
md['liver big'] = md['liver big'].astype(str).astype(int)
md['bilirubin'] = md['bilirubin'].astype(str).astype(float)

md.albumin.replace(0 , md.albumin.mean(), inplace= True)
md.bilirubin.replace(0 , md.bilirubin.mean(), inplace = True)
print(md.dtypes)
print(md.head())

# mini_dataset.shape
```

```python
md['sex'][md.sex == 1] = 'male'
md['sex'][md.sex == 2] = 'female'
md['antivirals'][md.antivirals == 1] = 'no'
md['antivirals'][md.antivirals == 2] = 'yes'
md['liver big'][md['liver big'] == 1] = 'no'
md['liver big'][md['liver big'] == 2] = 'yes'
md.head()
```
