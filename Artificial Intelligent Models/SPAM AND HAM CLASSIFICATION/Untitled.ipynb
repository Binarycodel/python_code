```python
import pandas as pd
df = pd.read_csv('data.csv')
print(df.head())
```

```python
df.info()
df.head(20)
```

```python
new_df = df[['Duration', 'Calories']]
new_df.head()
```

```python
mean = new_df['Calories'].mean()
dataset = new_df.fillna({
    'Calories': mean
})
# dataset.info()
# new_df.to_csv('data1.csv')
dataset.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
dura = dataset['Duration']
pls = dataset["Calories"] 
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.scatter(dura, pls, color='r' , marker='*')
```

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(dataset[['Duration']], dataset.Calories)
```

```python
y_prediction = lm.predict(dataset[['Duration']])
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
import pickle

with open('model_one', 'wb') as file:
    pickle.dump(lm, file)
    
with open('model_one' , 'rb') as file:
    lm = pickle.load(file)
    
    print(lm.predict([[350]]))
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Duration (time)')
plt.ylabel('Calories (energy)')
plt.scatterter(dataset.Duration, dataset.Calories, color="red", marker='*')
plt.plot(dataset.Duration, y_prediction, color="blue")
```

```python
import numpy as np 
import math
import matplotlib.pyplot as pt

x = np.array([1,2,3,4,5,6])
y = np.array([7,10,13,16,19,22])


def gradient_decent(x,y):
    learning_rate = 0.01
    m_init = 0 
    b_init = 0
    epoch = 1000
    n = len(x)
    
    for i in range(epoch):
        
        y_prediction = (m_init * x) + b_init
#   compute the cost of pridcition
        cost = (1/n) * sum([math.pow(val,2) for val in (y-y_prediction)])
        md = -(2/n) * sum(x*(y-y_prediction))
        bd = -(2/n) * sum((y-y_prediction))
        m_init = m_init - learning_rate  * md
        b_init = b_init - learning_rate * bd
        print(f'm [{m_init}] | b [{b_init}] |  cost [{cost}] | epoch = [{i}]')
    
gradient_decent(x,y)
    
```
